```python
import os 
import pandas as pd
import pickle
```

```python
df = pd.read_csv("../data/source/Colleges.csv")
```

```python
len(df)
```

```python
df.head()
```

```python
munged = df.dropna(axis=0, subset=['ProgramName',"Web"])
```

```python
len(munged)
```

```python
a = lambda: len(munged) * 100 // len(df) 
print(
    "The munged dataset is of size {} and has decrease of {}% from the original size of {}"
    .format(len(munged),100-a(),len(df))
)
```

```python
uniqueCount = len(munged["Name"].unique())
data = [munged[munged["Name"] == munged["Name"].unique()[e]].to_dict("records") for e in range(uniqueCount)]
```

```python
data = sum(data,[])
```

```python
pd.DataFrame(data).to_json(f"../data/full/college-tab.json",orient="records")
```
